```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#  3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
x1 = np.linspace(0, 4, 100)
x2 = np.linspace(0, 4, 100)
x1, x2 = np.meshgrid(x1, x2)

# 
z1 = (6 - x1 - 2 * x2) / 3  # x1 + 2x2 + 3x3 <= 6
z2 = (4 - 2 * x1 - x2)      # 2x1 + x2 + x3 <= 4
z3 = (5 - x1 - x2) / 2      # x1 + x2 + 2x3 <= 5

# 
z_feasible = np.minimum(np.minimum(z1, z2), z3)
z_feasible[z_feasible < 0] = np.nan  # 

# 
ax.plot_surface(x1, x2, z_feasible, alpha=0.5, color='blue')

# #  z = 
z_solution = (6 - x1 - 2 * x2) / 3   #  Z = 10 3x1 + 4x2 + 5x3
ax.plot_surface(x1, x2, z_solution, alpha=0.8, color='red')

# 
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.set_zlim(0, 4)

# 
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.set_title('3D Visualization of Feasible Region and Solution Plane')

plt.show()
```
